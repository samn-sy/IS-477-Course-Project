```python
import pandas as pd
```

```python
chicago = pd.read_csv("11.7_Chicago_Food_Inspections_Cleaned_Reduced.csv")
nyc = pd.read_csv("11.16_NYC_Inspections_Cleaned_Reduced.csv")
```

```python
chicago = chicago.rename(
    columns={"License #": "Restaurant ID"}
)
# convert Zip to Integer instead of Float
chicago['Zip'] = chicago['Zip'].astype('Int64')

chicago = chicago[["Restaurant ID", "DBA Name", "Risk", "Address", "City", "State", "Zip", "Inspection Date", "Violations", "Results"]]

chicago.head()
```

```python
nyc["City"] = "NEW YORK CITY"
nyc["State"] = "NY"

nyc = nyc.rename(
    columns={"CAMIS": "Restaurant ID", "DBA": "DBA Name", "BORO": "Boro", "STREET": "Street", "ZIPCODE": "Zip", "CUISINE DESCRIPTION": "Cuisine Description", "INSPECTION DATE": "Inspection Date", "VIOLATION DESCRIPTION": "Violations", "CRITICAL FLAG": "Critical Flag"}
)

nyc = nyc[["Restaurant ID", "DBA Name", "Boro", "Critical Flag", "Street", "City", "State", "Zip", "Cuisine Description", "Inspection Date", "Violations"]]

nyc.head()
```

```python
final_df = pd.concat([nyc, chicago], ignore_index=True)
final_df.index = range(1, len(final_df) + 1)
final_df['RecordID'] = final_df.index
final_df['Restaurant ID'] = final_df['Restaurant ID'].astype('Int64')

final_df
```

```python
final_df.to_csv('11.16_Inspections.csv', index=False)
```

```python
# Reducing for Github Repo
final_red = final_df.sample(frac = .75, random_state=477)
final_red
```

```python
final_red.to_csv('11.16_Inspections_Reduced.csv', index=False)
```
